## ¿Cuál es la categoría de producto que más ventas genera en total?

```python
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

## ¿Cuál es la media de ventas por Region y Category a la vez?

```python
df.groupby(["Region","Category"])["Sales"].mean()
```

## ¿Cuántos pedidos distintos hay por Region?

```python
df.groupby("Region")["Order ID"].nunique()
```

## Convertir 'Orden Date' de string a fecha

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
```

```python
df.dtypes
```

```python
df.shape
```

## ¿Cuáles son las ventas totales por año?


```python
df.groupby(df["Order Date"].dt.year)["Sales"].sum()
```

## ¿Cuántas ventas de media por cada mes del año?


```python
df.groupby(df["Order Date"].dt.month)["Sales"].mean().sort_values(ascending=False)
```

```python
df.groupby([df["Order Date"].dt.year, df["Order Date"].dt.month])["Sales"].sum()
```
